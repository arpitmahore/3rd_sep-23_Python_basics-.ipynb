17. Write a Python program to calculate the roots of a quadratic equation.

```python
import cmath

def quadratic_roots(a, b, c):
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    # Check if discriminant is positive, negative, or zero
    if discriminant > 0:
        root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        root2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        return root1, root2
    elif discriminant == 0:
        root = -b / (2*a)
        return root
    else:
        real_part = -b / (2*a)
        imaginary_part = cmath.sqrt(-discriminant) / (2*a)
        root1 = complex(real_part, imaginary_part)
        root2 = complex(real_part, -imaginary_part)
        return root1, root2

# Example usage:
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

roots = quadratic_roots(a, b, c)

if isinstance(roots, tuple):
    print(f"Root 1: {roots[0]}")
    print(f"Root 2: {roots[1]}")
else:
    print(f"Root: {roots}")
```

9. Create a program to calculate the total cost of a shopping cart based on discounts.

```python
def calculate_total_cost(items, discounts, tax_rate):

  subtotal = 0
  total_discount = 0
  total_tax = 0

  for item in items:
    subtotal += item["price"]

  for discount in discounts:
    total_discount += subtotal * discount["percentage"] / 100

  total_tax = (subtotal - total_discount) * tax_rate / 100

  # Calculate the total cost by adding the subtotal, the total tax, and subtracting the total discount
  total_cost = subtotal + total_tax - total_discount

  return total_cost

# Create a list of items with their names and prices
items = [
  {"name": "Apple", "price": 1.00},
  {"name": "Banana", "price": 0.50},
  {"name": "Orange", "price": 0.75},
  {"name": "Milk", "price": 3.00},
  {"name": "Eggs", "price": 2.50}
]

# Create a list of discounts with their names and percentages
discounts = [
  {"name": "Fruits", "percentage": 10},
  {"name": "Dairy", "percentage": 5}
]

# Set the tax rate as a percentage
tax_rate = 15

# Call the function and print the result
total_cost = calculate_total_cost(items, discounts, tax_rate)
print(f"The total cost of the shopping cart is ${total_cost:.2f}")
```

17. Write a program that calculates the average of a list of numbers, excluding the smallest and largest values.

```python
def average_excluding_min_max(lst):

  # Check if the list is empty or has only one element
  if not lst or len(lst) == 1:
    return None

  # Calculate the sum of the list
  total = sum(lst)

  # Subtract the smallest and largest values from the sum
  total -= min(lst)
  total -= max(lst)

  # Calculate the average by dividing the sum by the length of the list minus 2
  average = total / (len(lst) - 2)
  # Return the average
  return average



numbers = [11, 15, 9, 55, 41, 35, 20, 62, 49]

print(f"The average of the list, excluding the smallest and largest values, is {average_excluding_min_max(numbers)}")
```

20. Write a program that determines the roots of a cubic equation using the Cardano formula.

```python
import math

def cardano(a, b, c, d):
  # Check if the equation is degenerate
  if a == 0:
    return -d / b if b != 0 else 0

  # Calculate the discriminant
  delta = (b ** 2 - 3 * a * c) ** 0.5

  # Calculate the coefficients of the quadratic equation
  a2 = 3 * a
  b2 = 2 * b
  c2 = c

  # Calculate the roots of the quadratic equation
  x1 = (-b2 + delta) / a2
  x2 = (-b2 - delta) / a2

  # Calculate the root of the cubic equation
  x3 = (-c2 / x1 ** 2 - x1) / 3

  return x1, x2, x3

# Get the coefficients from the user
a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))
d = float(input("Enter the coefficient d: "))

# Calculate the roots of the cubic equation
x1, x2, x3 = cardano(a, b, c, d)

# Print the roots
print(f"The roots of the cubic equation are {x1}, {x2}, and {x3}.")
```

31. Create a program that validates a password based on complexity rules (length, characters, etc.)

```python
import re

def is_valid_password(password):
    # Check if the password is at least 8 characters long
    if len(password) < 8:
        return False

    # Check if the password contains at least one uppercase letter
    if not re.search('[A-Z]', password):
        return False

    # Check if the password contains at least one lowercase letter
    if not re.search('[a-z]', password):
        return False

    # Check if the password contains at least one number
    if not re.search('[0-9]', password):
        return False

    # Check if the password contains at least one special character
    if not re.search('[!@#$%^&*()-_=+]', password):
        return False

    # If all of the checks pass, then the password is valid
    return True

def main():
    # Get the password from the user
    password = input("Enter your password: ")

    # Validate the password
    if is_valid_password(password):
        print("Your password is valid.")
    else:
        print("Your password is not valid.")

if __name__ == "__main__":
    main()
```

35. Create a program that simulates a basic text-based tic-tac-toe game against the computer

```python
import random

board = [[' ', ' ', ' '],
         [' ', ' ', ' '],
         [' ', ' ', ' ']]

def print_board():
  for row in board:
    print(' | '.join(row))

def check_winner():
  for row in range(3):
    if board[row][0] == board[row][1] == board[row][2] != ' ':
      return board[row][0]
    if board[0][row] == board[1][row] == board[2][row] != ' ':
      return board[0][row]
    if board[row][0] == board[1][1] == board[2][2] != ' ':
      return board[row][0]
    if board[2][0] == board[1][1] == board[0][2] != ' ':
      return board[2][0]
  return None

def computer_move():
  while True:
    row = random.randint(0, 2)
    col = random.randint(0, 2)
    if board[row][col] == ' ':
      board[row][col] = 'O'
      break

def player_move():
  while True:
    row = int(input('Enter row (1-3): ')) - 1
    col = int(input('Enter column (1-3): ')) - 1
    if board[row][col] == ' ':
      board[row][col] = 'X'
      break

def main():
  print_board()
  while True:
    computer_move()
    print_board()
    winner = check_winner()
    if winner:
      print(f'The winner is {winner}!')
      break
    player_move()
    print_board()
    winner = check_winner()
    if winner:
      print(f'The winner is {winner}!')
      break

if __name__ == '__main__':
  main()
```

37. Develop a program that calculates the nth term of the Fibonacci sequence using memoization.


```python
def memoized_fibonacci(n):
  cache = {0: 0, 1: 1}

  def fib(n):
    if n in cache:
      return cache[n]
    else:
      cache[n] = fib(n - 1) + fib(n - 2)
      return cache[n]

  return fib(n)


def main():
  n = int(input("Enter the nth term of the Fibonacci sequence: "))
  print(f"The {n}th term of the Fibonacci sequence is {memoized_fibonacci(n)}.")


if __name__ == "__main__":
  main()
```

40. Write a program that generates the prime factors of a given number using trial division.v

```python
def prime_factors(n):
  factors = []
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      factors.append(i)
      while n % i == 0:
        n //= i
  if n > 1:
    factors.append(n)
  return factors

def main():
  n = int(input("Enter a number: "))
  print("The prime factors of", n, "are:", prime_factors(n))

if __name__ == "__main__":
  main()
```
